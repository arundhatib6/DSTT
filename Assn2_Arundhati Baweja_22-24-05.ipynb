```python
#required imports
import pandas as pd
import warnings
import io
warnings.filterwarnings('ignore')

from google.colab import files
 
 
uploaded = files.upload()
```

```python
#Read the housingdata.csv file into pandas DataFrame and display first six rows of the DataFrame
dataf = pd.read_csv(io.BytesIO(uploaded['housingdata.csv']))
print(dataf)
```

```python
#Display the column names and row index in separate cells.
column_names=list(dataf.columns)
column_names.insert(0,'')
new_dataf=pd.DataFrame(columns=column_names)
new_dataf=new_dataf.append(dataf)
new_dataf.reset_index(inplace=True)
new_dataf.rename(columns={'index':'Row Index'},inplace=True)
new_dataf
```

```python
#a) How many numbers of “STATE” exist in the dataset. Hint: use shape
#b) How many unique states exist in the dataset
print(dataf['STATE'].shape)
print(dataf['STATE'].unique())
```

```python
#Retrieve the list of all NaN/Null/Empty cells in the form of Boolean list
bool_dataf=dataf.isna()|dataf.isnull()|dataf.eq('')
print(bool_dataf)
```

```python
##Drop all rows with N/A,NA,na values in Num_Bedrooms
dataf=dataf[~dataf['NUM_BEDROOMS'].isin(['na','NaN','N/A'])]
print(dataf)
```

```python
#Replace the NaN and String entries in Num_Bath with previous entry
dataf['NUM_BATH']=pd.to_numeric(dataf['NUM_BATH'],errors='coerce')
dataf['NUM_BATH']=dataf['NUM_BATH'].fillna(method='ffill')
print(dataf)
```

```python
import numpy as np
```

```python
#Replace the empty values in “SQ_FT” with the mean of the all entries.
dataf['SQ_FT']=dataf['SQ_FT'].replace('--',np.nan)
dataf['SQ_FT']=dataf['SQ_FT'].astype(float)
mean_sqft=dataf['SQ_FT'].mean()
dataf['SQ_FT'].fillna(mean_sqft,inplace=True)
print(dataf)
```

```python
#What is the sum of SQ_FT of all entries in the dataset.
dataf['SQ_FT'].sum()
```

```python
#Add a new column “NUM_STORE” and with all the values as 1 for all rows
dataf['NUM_STORE']=[1,1,1,1,1,1,1,1]
dataf
```

```python
#Create a DataFrame containing all entries of TN state only
dataf_tn=dataf[dataf['STATE']=='TN']
dataf_tn=dataf_tn.reset_index()

dataf_tn.drop('index',axis=1,inplace=True)
dataf_tn
```

```python
#Create a DataFrame having SQ_FT area greater than 1000 and display the DataFrame
dataf1=dataf[dataf['SQ_FT']>1000.0]
dataf1=dataf1.reset_index()
dataf1.drop('index',axis=1,inplace=True)
dataf1
```

```python
#Display the state having highest average price per square foot of housing area.
state_prices=dataf.groupby('STATE')['PRICE','SQ_FT'].mean()
state_prices['pricepersqft']=state_prices['PRICE']/state_prices['SQ_FT']
state_prices=state_prices.sort_values('pricepersqft',ascending=True)
print(state_prices.index[0])
```
